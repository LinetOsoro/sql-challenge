```python
import pandas as pd
import os
```

```python
# Load the CSV files into pandas DataFrames
employees_df = pd.read_csv("Data/employees.csv")
employees_df.head()
```

```python
departments_df = pd.read_csv("Data/departments.csv")
departments_df.head()
```

```python
dept_manager_df = pd.read_csv("Data\dept_manager.csv")
dept_manager_df.head()
```

```python
dept_emp_df = pd.read_csv("Data\dept_emp.csv")
dept_emp_df.head()
```

```python
salaries_df = pd.read_csv("Data\salaries.csv")
salaries_df.head()
```

```python
titles_df = pd.read_csv("Data/titles.csv")
titles_df.head()
```

```python
# Merge employees and salaries based on 'emp_no'
emp_salaries_df = pd.merge(employees_df, salaries_df, on='emp_no')
emp_salaries_df.head(10) 
```

```python
# Merge dept_emp with departments based on 'dept_no'
emp_dept_df = pd.merge(dept_emp_df, departments_df, on='dept_no')
emp_dept_df.head(5)
```

```python
# Merge dept_manager with employees based on 'emp_no'
dept_manager_emp_df = pd.merge(dept_manager_df, employees_df, on='emp_no')
dept_manager_emp_df.head()
```

```python
# Merge dept_manager with employees based on 'emp_no'
dept_manager_emp_df = pd.merge(dept_manager_df, employees_df, on='emp_no')

# Merge dept_manager_emp_df with emp_dept_df based on 'emp_no' to add 'dept_name'
dept_manager_emp_df = pd.merge(dept_manager_emp_df, emp_dept_df[['emp_no', 'dept_name']], on='emp_no')
dept_manager_emp_df.head()
```

```python
# Merge the result from step 1 with titles based on 'emp_title_id'
emp_salaries_titles_df = pd.merge(emp_salaries_df, titles_df, left_on='emp_title_id', right_on='title_id')
emp_salaries_titles_df.head()
```

```python
#List the employee number, last name, first name, sex, and salary of each employee
# Constructing a DataFrame with specified columns
employees_info = emp_salaries_titles_df[['emp_no', 'last_name', 'first_name', 'sex', 'salary']]
employees_info.head(10)
```

```python
# List the first name, last name, and hire date for the employees who were hired in 1986

# Convert hire date column to datetime object
employees_df['hire_date'] = pd.to_datetime(employees_df['hire_date'])

# Filter employees hired in 1986
employees_hired_1986 = employees_df[employees_df['hire_date'].dt.year == 1986]

# Format hire date as '4/28/1986'
employees_hired_1986.loc[:, 'hire_date'] = employees_hired_1986['hire_date'].dt.strftime('%m/%d/%Y')


print(employees_hired_1986[['first_name', 'last_name', 'hire_date']])
```

```python
# List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B
hercules_b = employees_df[(employees_df['first_name'] == 'Hercules') & (employees_df['last_name'].str.startswith('B'))]

print(hercules_b[['first_name', 'last_name', 'sex']])
```

```python
# List each employee in the Sales department
sales_employees = dept_manager_emp_df[dept_manager_emp_df['dept_name'] == 'Sales']
print(sales_employees[['emp_no', 'last_name', 'first_name']])
```

```python
# List each employee in the Sales and Development departments
sales_dev_employees = dept_manager_emp_df[dept_manager_emp_df['dept_name'].isin(['Sales', 'Development'])]

print(sales_dev_employees[['emp_no', 'last_name', 'first_name', 'dept_name']])
```

```python
# List the frequency counts, in descending order, of all the employee last names
last_name_counts = employees_df['last_name'].value_counts().reset_index()
last_name_counts.columns = ['last_name', 'frequency']
last_name_counts = last_name_counts.sort_values(by='frequency', ascending=False)

print(last_name_counts)
```
